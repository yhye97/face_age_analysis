```python
import requests
import csv
from bs4 import BeautifulSoup

asian_actors=[]
asian_actors_list_name=[]#Asian actors
asian_actors_list_img=[];#asian actors img url
asian_actors_url={};
asian_actors_birth={};#asian actors birthday
asian_actors_age={}; #Asian actors age (2019)

req = requests.get('https://www.imdb.com/list/ls056023717/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
total=soup.select(
    '#main > div > div.lister.list.detail.sub-list > div.header.filmosearch > div > div.desc.lister-total-num-results'
)
total_num=""
for tag in total:
    total_num=str(tag.text).replace(" ","")
    total_num=total_num.replace("names","")
    break;

req = requests.get('https://www.imdb.com/list/ls056023717/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
names=soup.select(
        '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-content > h3 > a'
)

for name in names:
    asian_actors_dict=dict()
    asian_actors_dict['name']=str(name.text).replace("\n","").replace(" ","")
    asian_actors_url[str(name.text).replace("\n","").replace(" ","")]='https://www.imdb.com'+name.get('href')
    asian_actors_list_name.append(asian_actors_dict)

for key in asian_actors_url:
    req = requests.get(asian_actors_url[key])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    bday = soup.select(
    '#name-born-info'
    )
    for info in bday:
        if(info.select('h4')):
            if(info.select('time')):
                for date in info.select('time'):
                    asian_actors_birth[key]=date['datetime']
                    asian_actors_age[key]=2019-(int(str(date['datetime']).split('-')[0]))
    
#with open('asian_actors_bday.json', 'w+') as json_file:
#    json.dump(asian_actors_list_bday, json_file)
        
req = requests.get('https://www.imdb.com/list/ls056023717/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
img=soup.select(
    '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-image > a > img'
)

for src in img:
    no_image_url="https://m.media-amazon.com/images/G/01/imdb/images/nopicture/medium/name-2135195744._CB470041852_.png"
    if(src.get('src')!=no_image_url):
        asian_actors_dict=dict()
        asian_actors_dict['src']=str(src.get('src') )
        asian_actors_list_img.append(asian_actors_dict)
    else: #when there is no image
        asian_actors_dict=dict()
        asian_actors_dict['src']="" 
        asian_actors_list_img.append(asian_actors_dict)
        
#with open('asian_actors_name.json', 'w+') as json_file:
#    json.dump(asian_actors_list_name, json_file)
    
#with open('asian_actors_img.json', 'w+') as json_file:
#    json.dump(asian_actors_list_img, json_file)        

i=0
while i<int(total_num):
    list1=[]
    list1.append(asian_actors_list_name[i]['name'])
    list1.append(asian_actors_list_img[i]['src'])
    for key in asian_actors_birth:
        if(str(key) == asian_actors_list_name[i]['name']):
            list1.append(str(asian_actors_birth[key]))
            for k in asian_actors_age:
                if(str(k) == asian_actors_list_name[i]['name']):
                    list1.append(str(asian_actors_age[k]))
    asian_actors.append(list1)
    i=i+1
    
for index in asian_actors:
    if(len(index)!=4):
        asian_actors.remove(index)
        
with open("asian_actors.csv", 'w') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(asian_actors)
    
```
