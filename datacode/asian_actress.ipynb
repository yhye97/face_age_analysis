```python
import requests
import csv
import json
from bs4 import BeautifulSoup

asian_actress=[] #Asian actress name, img, bday 
asian_actress_list_name=[]#Asian actress name
asian_actress_list_img=[]#Asian actress img
asian_actress_url={}
asian_actress_birth={}
asian_actress_age={}; #Asian actress age (2019)

req = requests.get('https://www.imdb.com/list/ls053353032/?ref_=otl_1')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
total=soup.select(
    '#main > div > div.lister.list.detail.sub-list > div.header.filmosearch > div > div.desc.lister-total-num-results'
)
total_num=""

for tag in total:
    total_num=str(tag.text).replace(" ","")
    total_num=total_num.replace("names","")
    break;

req = requests.get('https://www.imdb.com/list/ls053353032/?ref_=otl_1')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
names=soup.select(
        '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-content > h3 > a'
)


for name in names:
    asian_actress_dict=dict()
    asian_actress_dict['name']=str(name.text).replace("\n","").replace(" ","")
    asian_actress_dict['url']='https://www.imdb.com'+name.get('href')
    asian_actress_url[str(name.text).replace("\n","").replace(" ","")]='https://www.imdb.com'+name.get('href')
    asian_actress_list_name.append(asian_actress_dict)

    
for key in asian_actress_url:
    req = requests.get(asian_actress_url[key])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    bday=soup.select(
    '#name-born-info'
    )
    for info in bday:
        if(info.select('h4')):
            if(info.select('time')):
                for date in info.select('time'):
                    asian_actress_birth[key]=date['datetime']
                    asian_actress_age[key]=2019-(int(str(date['datetime']).split('-')[0]))

            
req = requests.get('https://www.imdb.com/list/ls053353032/?ref_=otl_1')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
img=soup.select(
    '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-image > a > img'
)

for src in img:
    no_image_url="https://m.media-amazon.com/images/G/01/imdb/images/nopicture/medium/name-2135195744._CB470041852_.png"
    if(src.get('src')!=no_image_url):
        asian_actress_dict=dict()
        asian_actress_dict['src']=src.get('src')
        asian_actress_list_img.append(asian_actress_dict)
    else:
        asian_actress_dict=dict()
        asian_actress_dict['src']=""
        asian_actress_list_img.append(asian_actress_dict)

#with open('asian_actress_name.json', 'w+') as json_file:
#    json.dump(asian_actress_list_name, json_file)
    
#with open('asian_actress_img.json', 'w+') as json_file:
#    json.dump(asian_actress_list_img, json_file)        
        
i=0
while i<int(total_num):
    list1=[]
    list1.append(asian_actress_list_name[i]['name'])
    list1.append(asian_actress_list_img[i]['src'])
    for key in asian_actress_birth:
        if(str(key) == asian_actress_list_name[i]['name']):
            list1.append(str(asian_actress_birth[key]))
            for k in asian_actress_age:
                if(str(k) == asian_actress_list_name[i]['name']):
                    list1.append(str(asian_actress_age[k]))
    asian_actress.append(list1)
    i=i+1
    
for index in asian_actress:
    if(len(index)!=4):
        asian_actress.remove(index)
    else:
        print(index)
    
with open('asian_actress.csv', 'w', encoding='utf-8') as resultFile:
    writer = csv.writer(resultFile, dialect='excel')
    writer.writerows(asian_actress)
    
```
