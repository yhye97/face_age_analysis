```python
import requests
import csv
import json
from bs4 import BeautifulSoup

white_actors=[]
caucasian_celebs_urls=[]

f = open('caucasian_celebs2_url.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

for line in rdr:
    if (len(line)!= 0): #exclude empty elements
        if(line[0]!=''):
            caucasian_celebs_urls.append(line)
print(len(caucasian_celebs_urls))            
for url in caucasian_celebs_urls:
    white_actors_list_name=[]#white actors
    white_actors_list_img=[];#white actors img url
    white_actors_url={}; #white actors url
    white_actors_birth={}; #white actors birthday
    white_actors_age={}; #white actors age (2019)
    
    req = requests.get(url[0])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    names=soup.select(
        '#container > li > div > div > a'
    )

    for name in names:
        white_actors_dict=dict()
        white_actors_dict['name']=name.get('title')
        white_actors_url[str(name.get('title'))]=name.get('href')
        white_actors_list_name.append(white_actors_dict)
    print(len(white_actors_list_name))
    
    for key in white_actors_url:
        req = requests.get(white_actors_url[key])
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        bday = soup.select(
        '#wrap-body > div > div.posr > div > div > div.paratxt2 > p'
        )
        for info in bday:
            index=(str(info).find(')'))
            num=str(info)[index-4:index]
            if(num.isdigit()):
                white_actors_age[key]=2019-int(num)
            else:
                white_actors_age[key]=""
        
    req = requests.get(url[0])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    img=soup.select(
        '#container > li > div > div > a > img'
    )

    for src in img:
        white_actors_dict=dict()
        white_actors_dict['src']=str(src.get('src') )
        white_actors_list_img.append(white_actors_dict)
              
    i=0
    while i<len(white_actors_list_name):
        list1=[]
        list1.append(white_actors_list_name[i]['name'])
        list1.append(white_actors_list_img[i]['src'])
        for key in white_actors_age:
            if(str(key) == white_actors_list_name[i]['name']):
                list1.append(str(white_actors_age[key]))
        white_actors.append(list1)
        i=i+1
    
for index in white_actors:
    if(len(index)!=3):
        white_actors.remove(index)

with open("caucasian_celebs2.csv", 'w',encoding='utf-8') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(white_actors)
    
    
```
