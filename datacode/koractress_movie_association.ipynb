```python
import requests
import csv
import json
from bs4 import BeautifulSoup

asian_actress=[] #Asian actress name, img, bday 

urls=['http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=2&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=3&sdate=&edate=',
    'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=4&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=5&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=6&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=7&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=8&sdate=&edate=',
    'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=9&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=10&sdate=&edate=',
    'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=11&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=12&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=13&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=14&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=15&sdate=&edate=',
     'http://www.kfaa.kr/bbs/board.php?bo_table=actor&sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&page=16&sdate=&edate='
]

for url in urls:
    asian_actress_list_name=[]#Asian actress name
    asian_actress_list_img=[]#Asian actress img
    asian_actress_list_bday=[]
    asian_actress_list_age=[]
    asian_actress_url={}
    
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    img=soup.select(
        '#fboardlist > div.tbl_hea.d01.tbl_wr.ap > table > tbody > tr > td.img-column > a > img'
    )
    
    for src in img:
        no_image_url="http://www.kfaa.kr/bbs/board.php?bo_table=actor&amp;wr_id=576&amp;sca=%EC%97%AC%EC%9E%90%ED%9A%8C%EC%9B%90%EA%B2%80%EC%83%89&amp;page=5"
        if(src.get('src')!=no_image_url):
            asian_actress_dict=dict()
            asian_actress_dict['src']=src.get('src')
            asian_actress_list_img.append(asian_actress_dict)
        else:
            asian_actress_dict=dict()
            asian_actress_dict['src']=""
            asian_actress_list_img.append(asian_actress_dict)
    
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    names=soup.select(
        '#fboardlist > div.tbl_hea.d01.tbl_wr.ap > table > tbody > tr > td.bo_tr_item > a'
    )
    
    for name in names:
        asian_actress_dict=dict()
        asian_actress_dict['name']=name.get('title')
        asian_actress_list_name.append(asian_actress_dict)
    
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    bday=soup.select(
        '#fboardlist > div.tbl_hea.d01.tbl_wr.ap > table > tbody > tr > td.bo_tr_item > a > p.col-info.col-info-small'
    )

    for info in bday:
        asian_actress_birth=dict()
        asian_actress_birth['bday']=(str(info)[40:50])
        asian_actress_list_bday.append(asian_actress_birth)
        
        asian_actress_age=dict()
        if ((str(info)[40:44].isdigit())):
            asian_actress_age['age']=2019-(int(str(info)[40:44]))
        else:
            asian_actress_age['age']=""
        asian_actress_list_age.append(asian_actress_age)
        
    i=0
    while i<13:
        list1=[]
        list1.append(asian_actress_list_name[i]['name'])
        list1.append(asian_actress_list_img[i]['src'])
        list1.append(asian_actress_list_bday[i]['bday'])
        list1.append(asian_actress_list_age[i]['age'])
        asian_actress.append(list1)
        i=i+1
    
for index in asian_actress:
    for i in index:
        if (i==''):
            asian_actress.remove(index)
    

#print(asian_actress)
with open('kor_asian_actress.csv', 'w', encoding='utf-8') as resultFile:
    writer = csv.writer(resultFile, dialect='excel')
    writer.writerows(asian_actress)
    
```
