```python
import requests
import csv
import json
from bs4 import BeautifulSoup

white_actors=[]
white_actors_list_name=[]#white actors
white_actors_list_img=[];#white actors img url
white_actors_url={}; #white actors url
white_actors_birth={}; #white actors birthday
white_actors_age={}; #white actors age (2019)

req = requests.get('https://www.imdb.com/list/ls075115011/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
names=soup.select(
        '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-content > h3 > a'
)

for name in names:
    white_actors_dict=dict()
    white_actors_dict['name']=str(name.text).replace("\n","").replace(" ","")
    white_actors_url[str(name.text).replace("\n","").replace(" ","")]='https://www.imdb.com'+name.get('href')
    white_actors_list_name.append(white_actors_dict)

for key in white_actors_url:
    req = requests.get(white_actors_url[key])
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    bday = soup.select(
    '#name-born-info'
    )
    for info in bday:
        if(info.select('h4')):
            if(info.select('time')):
                for date in info.select('time'):
                    white_actors_birth[key]=date['datetime']
                    white_actors_age[key]=2019-(int(str(date['datetime']).split('-')[0]))
    
with open('white_actors_bday.json', 'w+') as json_file:
    json.dump(white_actors_list_bday, json_file)
        
req = requests.get('https://www.imdb.com/list/ls075115011/')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
img=soup.select(
    '#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div.lister-item-image > a > img'
)

for src in img:
    no_image_url="https://m.media-amazon.com/images/G/01/imdb/images/nopicture/medium/name-2135195744._CB470041852_.png"
    if(src.get('src')!=no_image_url):
        white_actors_dict=dict()
        white_actors_dict['src']=str(src.get('src') )
        white_actors_list_img.append(white_actors_dict)
    else: #when there is no image
        white_actors_dict=dict()
        white_actors_dict['src']="" 
        white_actors_list_img.append(white_actors_dict)
        
with open('white_actors_name.json', 'w+') as json_file:
    json.dump(white_actors_list_name, json_file)
    
with open('white_actors_img.json', 'w+') as json_file:
    json.dump(white_actors_list_img, json_file)        

i=0
while i<100:
    list1=[]
    list1.append(white_actors_list_name[i]['name'])
    list1.append(white_actors_list_img[i]['src'])
    for key in white_actors_birth:
        if(str(key) == white_actors_list_name[i]['name']):
            list1.append(str(white_actors_birth[key]))
            for k in white_actors_age:
                if(str(k) == white_actors_list_name[i]['name']):
                    list1.append(str(white_actors_age[k]))
    white_actors.append(list1)
    i=i+1
    
for index in white_actors:
    if(len(index)!=4):
        white_actors.remove(index)

with open("white_actors.csv", 'w',encoding='utf-8') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(white_actors)
    
    
    
```
