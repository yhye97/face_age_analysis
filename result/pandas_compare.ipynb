```python
import csv
import math
import pandas as pd                     
import matplotlib.pyplot as plt          # plotting
import numpy as np                       # dense matrices
from scipy.sparse import csr_matrix
import random

plt.close('all')
asians=[]
final_result_asians=[]

f = open('results_asian_actors.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

f = open('results_asian_actress.csv', 'r', encoding='utf-8')
rdr2 = csv.reader(f)

f = open('results_asian_celebs.csv', 'r', encoding='utf-8')
rdr3 = csv.reader(f)

for line in rdr:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_asians.append(line)
        
for line in rdr2:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_asians.append(line)

for line in rdr3:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_asians.append(line)

list1=['name', 'real_age', 'faceapi_age', 'dif']
final_result_asians.insert(0,list1)
            
with open("pandas_asians.csv", 'w', encoding='utf-8') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(final_result_asians)

data=pd.read_csv('pandas_asians.csv', sep=',', dtype='unicode')

caucasians=[]
final_result_caucasians=[]

f = open('results_white_actors.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

f = open('results_white_actress.csv', 'r', encoding='utf-8')
rdr2 = csv.reader(f)

f = open('results_caucasian_celebs.csv', 'r', encoding='utf-8')
rdr3 = csv.reader(f)

for line in rdr:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_caucasians.append(line)
        
for line in rdr2:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_caucasians.append(line)

for line in rdr3:
    if(len(line)==4):
        if(line!=['name','real_age','faceapi_age','dif']):
            final_result_caucasians.append(line)

list1=['name2', 'real_age2', 'faceapi_age2', 'dif2']
final_result_caucasians.insert(0, list1)
            
with open("pandas_caucasians.csv", 'w', encoding='utf-8') as resultFile:
    wr = csv.writer(resultFile, dialect='excel')
    wr.writerows(final_result_caucasians)

data2=pd.read_csv('pandas_caucasians.csv', sep=',', dtype='unicode')

df_col = pd.concat([data,data2], axis=1)

plt.figure();



df_col['real_age']=pd.to_numeric(df_col['real_age'])
df_col['dif']=pd.to_numeric(df_col['dif'])
df_col['real_age2']=pd.to_numeric(df_col['real_age2'])
df_col['dif2']=pd.to_numeric(df_col['dif2'])

data['real_age']=pd.to_numeric(data['real_age'])
data['dif']=pd.to_numeric(data['dif'])


data2['real_age2']=pd.to_numeric(data2['real_age2'])
data2['dif2']=pd.to_numeric(data2['dif2'])

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print("Asians Result : ")
    print(data)
    print()
    print("Caucasians Result : ")
    print(data2)


X_axis = ['Asians', 'Caucasians']
Y_axis = [str(data['real_age'].mean()) , str(data2['real_age2'].mean())]

xs = [i for i, _ in enumerate(Y_axis)]

plt.xticks([i for i, _ in enumerate(X_axis)], X_axis)

plt.bar(xs,Y_axis)

plt.title("Compare Real Age Average")
plt.xlabel('Race')
plt.ylabel('Real Age Difference')
plt.show()

X_axis = ['Asians', 'Caucasians']
Y_axis = [str(data['dif'].mean()) , str(data2['dif2'].mean())]

xs = [i for i, _ in enumerate(Y_axis)]

plt.xticks([i for i, _ in enumerate(X_axis)], X_axis)

plt.bar(xs,Y_axis)

plt.title("Compare Age Dif Average")
plt.xlabel('Race')
plt.ylabel('Face Api Age - Real Age')
plt.show()

print()
print("Asians: ")
print(data.shape)
print("Real Age Average: "+ str(data['real_age'].mean()))
print("Dif Age Average : " +str(data['dif'].mean()))
print("Caucasians:")
print(data2.shape)
print("Real Age Average: "+ str(data2['real_age2'].mean()))
print("Dif Age Average : " +str(data2['dif2'].mean()))

```
